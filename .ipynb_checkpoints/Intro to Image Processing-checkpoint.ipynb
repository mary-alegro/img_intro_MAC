# Spatial Filtering

### 1D Filtering

Create an artifical signal:

```python
step_signal = np.zeros(100)
step_signal[50:] = 1
plt.plot(step_signal)
```

Add noise to make it more realistic:

```python
np.random.seed(0)
noisy_signal = (step_signal + np.random.normal(0, 0.35, step_signal.shape))
plt.plot(noisy_signal)
```

Create an averaging *kernel* to smooth some of the noise. Apply it using the *convolution* operation:

```python
mean_kernel = np.full(3, 1/3)
smooth_signal = np.convolve(noisy_signal, mean_kernel, mode='valid')
print(mean_kernel)
```

Plot original and smoothed signals:

```python
plt.plot(smooth_signal, color='red',label='Smooth signal')
plt.plot(noisy_signal, color='blue',label = 'Noisy signal')
plt.legend(loc='lower right')
```

Now create a difference *kernel* to detect where are the signal changes:

```python
diff_kernel = np.array([1, 0, -1])
diff_signal = np.convolve(step_signal, diff_kernel, mode='valid')
```

Plot the result:

```python
plt.plot(diff_signal, color='red',label='Signal changes')
plt.plot(step_signal, color='blue',label = 'Step signal')
plt.legend(loc='lower right')
```

Now we can try applying our new filter to the noise signal:

```python
diff_signal_noisy = np.convolve(noisy_signal, diff_kernel, mode='valid')

plt.plot(diff_signal_noisy, color='red',label='Signal changes')
plt.plot(noisy_signal, color='blue',label = 'Noisy signal')
plt.legend(loc='lower right')
```

Turns out, it doesn't work so well. We also need to smooth the signal noise in order to get some result. It this case, we can combine both filters together using the *convolution*:

```python
comb_kernel = np.convolve(diff_kernel, mean_kernel, mode='full')
print(comb_kernel)
```

You will still see noise after applying the new filter but the response will peak at the edge location:

```python
comb_signal = np.convolve(noisy_signal, comb_kernel, mode='valid')

plt.plot(comb_signal, color='red',label='Signal changes - combined kernels')
plt.plot(noisy_signal, color='blue',label = 'Noisy signal')
plt.legend(loc='lower right')
```

### 2D Filtering

The same concepts can be applied to 2d images. 

First, let's crate a noisy version of our brain image:

```python
noisy_img = (brain_img + (np.random.normal(0, 0.35, brain_img.shape)/255.))
plt.imshow(noisy_img, cmap='gray')
```

Let's manually create an mean *kernel*:

```python
mean_kernel = np.full((3, 3), 1/9)

print(mean_kernel)
```

```python
import scipy.ndimage as ndi #let's use scipy's convolution veersion since it handles image edges a little better.

mean_img = ndi.convolve(noisy_img, mean_kernel)

fig, ax = plt.subplots(1, 3, figsize=(15, 15))
ax[0].imshow(brain_img, cmap='gray')
ax[0].set_title('Original image');
ax[1].imshow(noisy_img, cmap='gray')
ax[1].set_title('Noisy image')
ax[2].imshow(mean_img, cmap='gray')
ax[2].set_title('Averaged image');
```

Now let's try to do some edge detection. For that we will create a set of vertical and horizontal kernels:

```python
vertical_kernel = np.array([
    [1, 1, 1],
    [0, 0, 0],
    [-1, -1, -1]
])
horizontal_kernel = vertical_kernel.T
```

We then convolve our image with our kernels:

```python
vertical_grad = ndi.convolve(brain_img,vertical_kernel)
horizontal_grad = ndi.convolve(brain_img,horizontal_kernel)

fig, ax = plt.subplots(1, 2, figsize=(10, 10))
ax[0].imshow(vertical_grad, cmap='gray')
ax[0].set_title('Vertical gradient');
ax[1].imshow(horizontal_grad, cmap='gray')
ax[1].set_title('Horizontal gradient')
```

And compute the magnitude of the filters responses, which gives us the edges:

```python
edges = np.sqrt(vertical_grad**2 + horizontal_grad**2)
plt.imshow(edges, cmap='gray')
```

# Histograms

Let's grab a fresh slice from our MRI data:

```python
slice_img = mri_vol[50,:,:]
plt.imshow(slice_img, cmap='gray')
plt.colorbar()
```

Our original image is *int16* but I want to convert it to a *uint8* to keep the histogram simple. Because *int16* variables can hold large values than *uint8*, the right way to do this conversion is first rescale the values to the 0..255 range and then convert the variable to a different type. This way we avoid clipping values and generating wrong results:

```python
tmp_img = slice_img/float(slice_img.max())
slice_img = tmp_img*255
brain_img = np.ubyte(brain_img)
```

Rotate image using Numpy, just for a change: 

```python
brain_img = np.rot90(slice_img)
plt.imshow(brain_img, cmap='gray')
```

```python
%matplotlib notebook
fig, ax = plt.subplots(1, 1)
ax.hist(brain_img.ravel(), bins=256, range=[0, 256])
ax.set_xlim(0, 256);
```

### Histogram Thresholding

You can use the histogram plot to check which values are good for manual thresholding. This is the most basic ways of segmenting an image:

```python
%matplotlib inline

segment = brain_img > 12
plt.imshow(segment*255, cmap='gray')
```

There are several histogram-based segmentation algorithms. One of the most popular ones in know as the Otsu method:

```python
import skimage.filters as filters

thres_otsu = filters.threshold_otsu(brain_img)
segment_otsu = (brain_img > thres_otsu)*255
plt.imshow(segment_otsu, cmap= 'gray')
```

Another popular one is the Triangle method:

```python
thres_tri = filters.threshold_triangle(brain_img)
segment_tri = (brain_img > thres_tri)*255
plt.imshow(segment_tri, cmap= 'gray')
```

*Scikit-image* has several other histogram segmentation algorithms. There is no hard rule on how to choose one algorithm for a especific task. Different algorithms will perform differently dipending on the type of image you are using. That's when image segmentation also becomes a craft. ;)

```python
import skimage.exposure as exp
%matplotlib inline
```

Another traditional use of image histogram is for image enhancement. For instance, we can use the histogram to equalize image intensities:

```python
brain_eq = exp.equalize_hist(brain_img)

fig, ax = plt.subplots(1, 2, figsize=(10, 10))
ax[0].imshow(brain_img, cmap='gray')
ax[0].set_title('Original image');
ax[1].imshow(brain_eq, cmap='gray')
ax[1].set_title('Equalized image')
```

Histogram equalization boosts the overall signal making background noise evident. We can use one of our masks from above to remove background pixels:

```python
brain_eq[segment_otsu == 0] = 0

fig, ax = plt.subplots(1, 2, figsize=(10, 10))
ax[0].imshow(brain_img, cmap='gray')
ax[0].set_title('Original image');
ax[1].imshow(brain_eq, cmap='gray')
ax[1].set_title('Equalized image')
```

Contrast-limited adaptive histogram equalization (CLAHE) is a more advanced histogram equalization algorithm that tends to generate better results:

```python
brain_eq2 = exp.equalize_adapthist(np.uint8(brain_img))

fig, ax = plt.subplots(1, 2, figsize=(10, 10))
ax[0].imshow(brain_img, cmap='gray')
ax[0].set_title('Original image');
ax[1].imshow(brain_eq2, cmap='gray')
ax[1].set_title('Equalized image')
```

# Feature Extraction

Convert equalized image to *uint8*:

```python
from skimage import img_as_ubyte
brain_eq2_8bits = img_as_ubyte(brain_eq2)
```

Extract pixel values from brain and 'neck' regions:

```python
#based on the scikit-image texture analysis tutorial
from skimage.feature import greycomatrix, greycoprops

PATCH_SIZE = 7

brain_locations = [(101, 151), (88, 99), (91, 126)]
brain_patches = []
for loc in brain_locations:
    brain_patches.append(brain_eq2_8bits[loc[0]:loc[0] + PATCH_SIZE,
                               loc[1]:loc[1] + PATCH_SIZE])

neck_locations = [(240, 135), (217, 138), (236, 65)]
neck_patches = []
for loc in neck_locations:
    neck_patches.append(brain_eq2_8bits[loc[0]:loc[0] + PATCH_SIZE,
                             loc[1]:loc[1] + PATCH_SIZE])
```

Plot location of hand picked patches:

```python
fig = plt.figure(figsize=(15, 15))
ax = fig.add_subplot(3,2,1)
ax.imshow(brain_eq2_8bits, cmap='gray')
for (y, x) in brain_locations:
    ax.plot(x + PATCH_SIZE / 2, y + PATCH_SIZE / 2, 'gs')
for (y, x) in neck_locations:
    ax.plot(x + PATCH_SIZE / 2, y + PATCH_SIZE / 2, 'bs')
ax.set_xlabel('Original Image')
ax.set_xticks([])
ax.set_yticks([])
ax.axis('image')
plt.show()
```

Compute texture features from brain and 'neck' image patches:

```python
xs = []
ys = []
for patch in (brain_patches + neck_patches):
    glcm = greycomatrix(patch, [5], [0], 256, symmetric=True, normed=True)
    xs.append(greycoprops(glcm, 'dissimilarity')[0, 0])
    ys.append(greycoprops(glcm, 'correlation')[0, 0])
```

Plot texture coefficients:

```python
fig = plt.figure(figsize=(15, 15))
ax = fig.add_subplot(3, 2, 2)
ax.plot(xs[:len(brain_patches)], ys[:len(brain_patches)], 'go',
        label='Brain')
ax.plot(xs[len(neck_patches):], ys[len(neck_patches):], 'bo',
        label='Neck')
ax.set_xlabel('GLCM Dissimilarity')
ax.set_ylabel('GLCM Correlation')
ax.legend()
plt.show()
```
